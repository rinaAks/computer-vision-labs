```python
# импортирование библиотек
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("cat.jpg") # это стандартное rgb-изображение
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# imshow показывает в новом окне
# cv2.imshow('RGB image', img)
# cv2.waitKey(0)

plt.title('RGB image')
plt.imshow(img)
```

```python
imgYUV = cv2.cvtColor(img, cv2.COLOR_RGB2YUV) # YUV-изображение

plt.title('YUV image')
plt.imshow(imgYUV)
```

```python
imgRGB = cv2.cvtColor(imgYUV, cv2.COLOR_YUV2RGB) # снова rgb-изображение

plt.title('RGB image after YUV')
plt.imshow(imgRGB)
```

```python
imgHSV = cv2.cvtColor(img, cv2.COLOR_RGB2HSV) # HSV-изображение

plt.title('HSV image')
plt.imshow(imgHSV)
```

```python
imgRGB = cv2.cvtColor(imgHSV, cv2.COLOR_HSV2RGB) # снова rgb-изображение

plt.title('RGB image after HSV')
plt.imshow(imgRGB)
```

```python
imgRGB2 = np.asarray(Image.open("cat.jpg"))
```

```python
plt.figure(figsize=[10, 5])

# Выводим на графиках r, g, b
b,g,r = cv2.split(imgRGB2)
plt.figure(figsize=[10, 5])

plt.subplot(1,3,1)
plt.imshow(r, cmap='gray')
plt.title("R")

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title("G")

plt.subplot(1,3,3)
plt.imshow(b, cmap='gray')
plt.title("B")
```

```python
# Выводим на графиках y, u, v
y,u,v = cv2.split(imgYUV)
plt.figure(figsize=[10, 5])

# яркость Y (Luma) 
plt.subplot(1,3,1)
plt.imshow(y, cmap='gray')
plt.title("Y")

# два цветоразностных компонента UV (Chroma)
plt.subplot(1,3,2)
plt.imshow(u, cmap='gray')
plt.title("U")

plt.subplot(1,3,3)
plt.imshow(v, cmap='gray')
plt.title("V")
```

```python
# Выводим на графиках h, s, v
h,s,v = cv2.split(imgHSV)
plt.figure(figsize=[10, 5])

# Цветовой тон (Hue)
plt.subplot(1,3,1)
plt.imshow(h, cmap='gray')
plt.title("H")

# Насыщенность (Saturation)
plt.subplot(1,3,2)
plt.imshow(s, cmap='gray')
plt.title("S")

# Яркость (Value, или Brightness)
plt.subplot(1,3,3)
plt.imshow(v, cmap='gray')
plt.title("V")
```
